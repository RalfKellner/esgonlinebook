# ESG rating divergence

[Berg et al. (2022)](https://academic.oup.com/rof/article/26/6/1315/6590670) compare ESG ratings of the following agencies:

* Kinder, Lydenberg and Domini (KLD)
* Sustainalytics
* Moody's ESG
* S\&P Global 
* Refinitiv
* MSCI

They quantify the ESG rating divergence of these agencies, but, their major contribution is to map the different approaches to a common taxonomy and decompose rating divergence into three components:

* scope
* measurement
* weight

Their results show that measurement (56\%) and scope (38\%) play the major role for rating divergence, while different weighting of ESG indicators only plays a minor role (6\%). The table below is from their paper and shows descriptive statistics for the agencies rating. Note that their analysis is based on ratings from 2014 and backed up by data from 2017.

## Sources for rating divergence

Berg et al. (2022) develop a common taxonomy to compare ESG ratings by the agencies. A direct comparison is not possible, because rating agencies use different structures to assess the rating. The structure differs in hierarchy as well as the use of indicators for quantifying different categories of ESG risks. With the developed common taxonomy, it is possible to determine differences in scope (the sole use of the ignoring of categories by sub-groups of agencies), in measurement (use of different indicators for the same category) and in weighting (different weights for aggregation of lower levels).

### Scope 

Berg et al. (2022) start with a union of 709 indicators which are used by rating agencies. According to the descriptions of these indicators they are divided into categories. Each indicator is exclusively assigned to one category and a new category is created if at least two indicators from different agencies describe the same attribute, e.g., water, corruption or diversity. Overall, 64 categories result and reveal that certain categories are not captured by all agencies. This demsonstrates differences in the scope of ESG ratings by these agencies.

### Measurement

In a next step, Berg et al. (2022) determine category scores for each rating agency by taking the average of indicators in each category. After this, pairwise category score correlations are determined between all agencies. Again, correlation values are rather low which demonstrates that agencies quantify the same ESG attributes in different ways

### Weighting

After normalizing all category scores to zero mean and unit variance, ESG ratings are regressed on category scores for each rating agencies. The size of estimated beta coefficients enables a comparison regarding the importance for each rating agency's assessment. Again, differences among agencies can be identified.

Summing this up, we observe that agenies determine their ratings with at least partly different attributes, different ways to measure these attributes and different weightings of these attributes. 

### Decomposition

Berg et al. (2022) decompose pairwise rating divergences by comparing estimtad rating scores on the level of categories which are exclusively used by one of the two agencies (scope), by quantiyfing the difference score based on common category scores and common weightings (measurement) and by quantifying the remainder as weighting difference.

In addition to the quantification of these effects, Berg et al. (2022) further identify a rater effect which relates to a bias that agencies are prone to by correlations between their own category scores. I.e., firms tend to be evaluated better (worse) in one category, if they perform good (bad) in another category. This further drives divergence among ESG ratings.

Overall, the paper demonstrates why ESG ratings and highlights the need for more unified standards to measure ESG performance among rating agencies.
