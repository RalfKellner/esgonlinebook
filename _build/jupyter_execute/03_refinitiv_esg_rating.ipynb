# Refinitiv ESG data

Refinitiv's description of its methodology is disclosed [here](https://www.refinitiv.com/content/dam/marketing/en_us/documents/methodology/refinitiv-esg-scores-methodology.pdf).

## Summary

* covers 85\% of market capitalization with its ratings
* uses 630 indicators
* data starts at 2002
* measures relative ESG performance with respect to sector and country
* Refinitiv claims that "it does not presume to define what good looks like; it lets the data determine industry-based relative performance..."
* they determine scores for four pillars: environmental, social, governance and controversy; the scores are aggregated to an ESG and an ESGC score

## Data

* they employ about 700 analysts to collect ESG data
* they collect data from publicly available sources

```python
from IPython.display import Image
Image(filename='refinitiv_data_sources.jpg', width = 100)
```

* the 630 indicators are manually processed by analysts
* the goal is to standardize each measure for comparability among all companies
* data is updated regularly on a yearly basis, controversies are included "on the fly"
* currently 12,500 companies with an emphasis on North America (3800), Asia (3100) and Europe (2500) 

## ESG scores

* a subset of 186 ESG measures is used 
* these measures are grouped into 10 categories:
    * Environmental: ressource use, emissions, innovation
    * Social: workforce, human rights, community, product responsibility
    * Governance: management, shareholders, csr strategy
* a score is determined for each category 
* each ESG pillar score is a relative sum of category weights which vary per industry for the environmental and the social pillar
* the ESGC score is accompanied by significant media controversies of the company and is supposed to discount the ESG score in case of large controversies
* 23 controversy topics exist which are aggregated to a controversy score
* Boolean ESG indicators are represented by 1 and 0, while 0 is the default value even if no information is available w.r.t. this indicator; numeric indicator variables are only used as reported and enter the scoring by a industry relative ranking; all indicators have a polarity which indicates if high values are positive or negative
* A ranking score for each of the 10 categories is determined by:

$$
\text{score} = \frac{\text{no. of companies with a worse value} + \frac{\text{no. of companies with the same value}}{2}}{\text{no. of companies with a value}}
$$

* The enironmental and social scores are determined on TRBC industry group level
* The governance score is determined based on country level
* Materiality of themes for each of the 10 categories and each industry is approximated by sufficient disclosure of data w.r.t. to the theme
* Category weights are determined based on materiality, see the documentation for more details
