# Environmental, social and corporate governance

Originally, the value and risk of companies mainly has been assessed by financial factors, e.g., balance sheet, financial, income or cash flow statements. These numbers are used to assess the future income flows and risks of a company. Nowadays, so called non-financial factors are more and more included into the company evaluation process which extends the analysis to the question how the company operates with respect to environmental, social and corporate governance (ESG) standards. Over time, the topic ESG has become more important and popular. For instance, the cell below shows the google trend search over the past 20 years.

In addition, ESG information is more and more included into investment decisions. For instance, according to the [2023 annual report](https://www.unpri.org/annual-report-2023) of the Principles for Responsible Investments more than 5,000 professional investors with far more than 100 trillion US Dollars assets under management have commited themselves to include ESG information in their investement decisions. This creates a demand for ESG ratings and is likely one of the reasons why we observe an increasing number of rating agencies offering ESG ratings. However, a huge challenge is the well documented inconsistency of ratings provided by these agencies. ESG rating inconsitencies point to the need for ESG standards and a common taxonomy which is important to reduce ESG uncertainty and its potential negative consequences. Thus, the aim of this course is to gain a deeper understanding for the ESG topic, hereby, putting an emphasis on the ESG rating inconsistency.

##  Definition of ESG

In contrast to measurable company events such as the default of a company, the failure or non-failure of ESG standards is not unambiguously defined. Even worse, two perspectives are common when it comes to the ESG defintion. (1) ESG quality measures the company's impact on the enironment, employees, customers and its community, and, (2) ESG quality quantifies the impact of environmental, social and governance factors on the company's business and operations. The latter point of view is more common for ESG rating providers.

Both defintions must not necessarily rule out similar conclusion of decision makers, but, also do not ensure consistency. Under (1), companies can improve their ESG quality by direct actions which improve the company's environment. Under (2), ESG factors serve as an additional source for strategic planning in order to cope with future developments which may harm the company's business and operations. 

Accordingly, conflicts arise if investors assume (1) and receive ratings according to (2), which is usually the case as ESG rating agencies mostly follow (2). Easily spoken, (1) measures the impact of a company on the world, and, (2) measure the impact of the world on the company.

## ESG rating dimensions

Rating providers first build ratings or scores on individual dimensions, e.g., environmental, social and governance scores. The final ESG rating is an aggregated score of its components, however, the weighting of each component is set by the agency (also on lower levels of aggregation). In addition, rating agencies tend to use subcomponents for each component and aggregates these for component scores. All scores are build upon historical data and may further be accompanied with theoretical reasoning and expert judgement. When it comes to the statistical analysis of historical data, agencies use a large set of partly different indicators. Another important note is the differentiation between absolute and relative ESG performance. Some rating agencies either build industry specific agencies which point towards ESG quality in relation to peer companies. However, other agencies try to provide a general point of view by assessing the absolute ESG quality. 

## ESG rating agencies

ESG rating agencies started to emerge a few decades ago, however, they gained most attraction in the past years. ESG ratings are often used to confirm certain quality standards which is popular for sustainable investing. Many ESG rating agencies are financial data providers, let us take a look at some examples.

* Refinitiv: rebranded from Thomson Reuters, it has been purchased by the London Stock Exchange Group in 2021 and offers ESG ratings for approximately 12,500 companies. We have access to these ratings with the university's license
* MSCI: originally a part of Morgan Stanley (which are well known for building stock market indexes), they became a serparately traded company which bought RiskMetrics whose subsidiary KLD has been one of the earliest data provider for US sustainability data; MSCI provides ESG ratings for about 8,500 companies
* Sustainalytics: owned by Morningstar which provides ratings for mutual and exchange-traded funds; Sustainalytics provides ESG ratings for about 13,000 which are at least in aggregated form freely accessible for most companies
* ISS: a subsidiary of Institutional Shareholder Services which provides recommendations to investment management firms; ISS provides ESG ratings for about 12,000 companies; Since 2020, Deutsche Brse owns 80\% of ISS

### Scope of ESG ratings

We will see later in more detail that rating agencies have a different scope of a company's quantified ESG quality, however, most rating agencies relate ESG quality and investment risk. The basic assumption is that companies may face business drawbacks and higher risks when they fail to meet ESG standards (in the future). Some agencies even claim to construct their ESG ratings such that they are supposed to be of predictive nature for future stock asset return performance. At the same time, other agencies additionally claim that their ratings give an assessment for the company's ESG commitment, its social impact and its transparency. 

### How rating agencies quantify their ratings

Rating providers either use a credit rating alike system, e.g., A+ (best) to D- (worst) or percentiles/scores, i.e., a value between 0 and 100. Here, 0 and 100 can mean very good or very bad performance, respectively. Either way, both measurements are ordinal scales which makes rank-based correlations the best choices for assessing rating correlation among different agencies. 

### Which data is used by rating agencies

Until today, ESG reporting standarads are in their infancy which is why rating agencies collect their own ESG data from a broader spectrum of data sources. Examples are public financial reports, news, voluntary ESG reports or answers from solicited questionnaires. See the links below which give you examples for Apple Inc.

* [Public 10-K report to the SEC](https://www.sec.gov/ix?doc=/Archives/edgar/data/320193/000032019322000108/aapl-20220924.htm)
* [Voluntary ESG report](https://s2.q4cdn.com/470004039/files/doc_downloads/2022/08/2022_Apple_ESG_Report.pdf)
* [News example](https://www.apple.com/newsroom/2022/10/apple-calls-on-global-supply-chain-to-decarbonize-by-2030/) 

Due to the lack of standardization, available data among companies is very different leading to incomplete structured datasets. Additionally, if companies report for the same topic, they may use different variables or scales.

### Impact of ESG ratings

Unfortunately, ESG ratings usually have not the impact they claim to have. ESG funds sometimes invest more in companies with greater failure with labor and environmental laws than non-ESG funds [(Raghunandan and Rajgopal, 2022)](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=3826357). Investors which sign the Principles for Responsible Investment seem not to invest in companies with better ESG performance [(Gibson et al., 2022)](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=3525530). Furthermore, a large survey study by [(Atz et al., 2022)](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=3708495) shows that ESG investing has been indistinguishable from other investment strategies.

## Regulation

While most of ESG regulation is not mandatory nor harmonized on a global level, ongoing developments head towards clear ESG disclosure rules. In the US, the Securities and Exchange Commission (SEC) requires publicly listed companies to disclose information that may be material to investors. This includes risks which are related to ESG factors. More recently, the SEC proposed the [enhancement and standardization of climate-related disclosure](https://www.sec.gov/files/33-11042-fact-sheet.pdf). This [video](https://youtu.be/-V12MqwVHuA?si=XjIA3ysYGefTxliA) makes it very clear that the SEC aims for standardization of ESG information to improve ESG assessment from an investor's perspective. To the best of my knowledge, SEC rules are the only mandatory framework of ESG disclosure in the US. However, voluntary international standards exist, e.g., the [Global Reporting Initiative](https://www.globalreporting.org/), the [Sustainability Accounting Standards Board](https://sasb.org/), [the Task Force on Climate-Related Financial Disclosure](https://www.fsb-tcfd.org/) and [the International Sustainability Standard Board](https://www.ifrs.org/groups/international-sustainability-standards-board/).

The EU and the UK already adopted mandated rules which affect ESG disclosure. The Fincancial Conduct Authority in the UK confirmed rules for climate-related disclosure in 2021 which came into play in 2022. According to these rules, issuers of listed equity shares or shares represented by certificates must include statements in the annual financial reports setting out whether their disclosures meet the recommendations of the Taskforce on Climate-related Financial Disclosures (TCFD). If they dont, theyll need to explain why. FCA-regulated asset managers and asset owners - including life insurers and pension providers - will have to disclose how they take climate-related risks and opportunities into account in managing investments. Theyll also have to make disclosures about the climate-related attributes of their products.

In the EU, companies with more than 500 employees must disclose non-financial and diversity information under the Non-Financial Reporting Directive (NFRD, Directive 2014/95/EU). However, great flexibility is given to these companies in terms of how they report this information. The Corporate Sustainability Reporting Directive (CSRD)  extends the and looks to address key structural weaknesses in current ESG regulation reporting. More companies (appr. 50,000 instead of 12,000) need to disclose ESG information and the scope of reporting is broadened. The CSRD rules apply to the financial year 2024 and will be reported first in 2025. Further important EU regulations are the Taxonomy Regulation (uniyfing ESG taxonomy), the Sustainable Finance Disclosure Regulation (transparency in sustainable investment) and the Corporate Due Dilligence Directive (minimze human rights and environmental risks). See this [document](https://www.squirepattonboggs.com/-/media/files/insights/publications/2023/07/esg-laws-across-the-world/esg-laws-across-the-world.pdf?rev=995db37cd01d448da71ab7853ff25867) for more ESG regulation information worldwide.
